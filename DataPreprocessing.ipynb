```python
def standardprice(price):
  if price[1] == 'Crore':
    return round(float(price[0]),2)
  elif price[1] == 'Lac':
    return round((float(price[0])/100),2)
```

```python
data['price'] = data['price'].str.split(' ').apply(standardprice)
```

```python
data.rename(columns={'price':'price(in Crore)'},inplace=True)
```

```python
data.head()
```

```python
data['price_per_sqft'] = data['price_per_sqft'].str.split('per').str.get(0).str.strip('₹').str.strip(' ').str.replace(',','').astype('float')
```

```python
data.head()
```

```python
data.info()
```

```python
data['bedRoom'].value_counts()
```

```python
#data['bedRoom'] = data['bedRoom'].str.split(' ').str.get(0).astype('int')
data['bathroom'] = data['bathroom'].str.split(' ').str.get(0).astype('int')
data['balcony'] = data['balcony'].str.split(' ').str.get(0).astype('int')
```

```python
data['balcony'] = data['balcony'].str.split(' ').str.get(0).str.replace('No','0')
```

```python
data['balcony'].value_counts()
```

```python
data['additionalRoom'].isnull().sum()
```

```python
data['additionalRoom'] = data['additionalRoom'].str.lower()
```
